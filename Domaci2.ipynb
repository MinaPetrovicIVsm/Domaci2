```python
import pandas as pd
import numpy as np
import math
import operator
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Real estate.csv")
data.head()
```

```python
X= data['X2 house age'].values
Y= data['Y house price of unit area'].values

mean_x=np.mean(X)
mean_y=np.mean(Y)
```

```python
s1=0
s2=0
for i in range(414):
  s1+=(X[i]- mean_x)*(Y[i]-mean_y)
  s2+=(X[i]-mean_x)**2

k=s1/s2
n=mean_y-(k*mean_x)

print (f'k = {k} \nn = {n}')
```

```python
x_max=np.max(X)
x_min=np.min(X)

x=np.linspace(x_min,x_max)
y = k*x+n

plt.plot(x,y,color='yellow', label='Regression line')
plt.scatter(X,Y,label='data')
plt.xlabel('latitude')
plt.ylabel('house price of unit area')
plt.show()
```
